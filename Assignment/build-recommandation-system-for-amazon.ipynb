```python
import numpy as np
import pandas as pd 
```

```python
data = pd.read_csv("1429_1.csv")
data
```

```python
df = data[['id','name','categories','reviews.rating','reviews.username']]
df
```

```python
df.isnull().sum()
```

```python
df.columns = ['product_id','product_name','categories','rating','username']
```

```python
product_details = df[['username','product_name','categories']]
product_details = product_details.drop_duplicates(subset = ['username'])
product_details
```

```python
df = df.dropna(subset=['rating'])
df
```

```python
product_data = pd.DataFrame(df.groupby('username')['rating'].mean())
product_data['Number of ratings'] = df.groupby('username')['rating'].count().values
product_data
```

```python
product_data.describe()
```

```python
product_data = product_data[product_data['Number of ratings'] >= 20]
product_data
```

```python
matrix_df = df[['username','rating','product_id']]
matrix_df
```

```python
matrix_df = matrix_df[matrix_df['username'].isin(product_data.index)]
matrix_df
```

```python
pivot_df = matrix_df.pivot_table(index='product_id',columns='username',values='rating').fillna(0)
pivot_df
```

```python
def recommend(prod_id):
    #the data set im using is very large so Im limiting the algorithm to only use 1000 rows of it
    #you can increase the amount of products that pop up as recommended by increasing the number
    #of rows included
    recomm_products = pivot_df.corrwith(pivot_df[prod_id]).sort_values(ascending=False)[:2000]
    t_data = product_details[product_details['username'].isin(recomm_products.index)].copy()
    t_data['corr'] = t_data['username'].apply(lambda x : recomm_products[x])
    t_data = t_data.sort_values(by='corr',ascending=False)
    return t_data

recommend("Kevin").drop_duplicates('product_name')
```
