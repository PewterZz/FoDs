```python
conditions=[(df['Age'] >= 55),(df['Age'] < 55) & (df['Age'] > 40),(df['Age'] <= 40)]
values = ['Elderly', 'Middle Aged', 'Young']
df['Group'] = np.select(conditions, values)
df
```

```python
elderly = 0
middle = 0
young = 0

for i in df["Group"]:
    if i == 'Elderly':
        elderly += 1
    elif i == 'Middle Aged':
        middle += 1
    elif i == 'Young':
        young +=1

plt.bar('Elderly', elderly, color='red', width=0.5)
plt.bar('Middle aged', middle, color='red', width=0.5)
plt.bar('Young', young, color='red', width=0.5)
plt.show()
```

```python
kant = pd.get_dummies(df)
```

```python
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(kant)
print(scaler.transform(kant))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
X_train, X_test, y_train, y_test = train_test_split(kant,df['AHD'], random_state= 1)

svm = SVC(kernel='rbf', random_state = 1, gamma = 0.5, C = 1)
svm.fit(X_train, y_train) 
predict = svm.predict(X_test)
print(accuracy_score(y_test,predict))
```

```python
from sklearn.metrics import classification_report
target_names = ['Yes','No']
print(classification_report(y_test, predict, target_names=target_names))
```
